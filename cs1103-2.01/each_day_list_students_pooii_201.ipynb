```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

from ipysheet import from_dataframe, to_dataframe
from ipywidgets import Text, VBox, link
```

```python
data = pd.read_csv("list_students_pooii_201.csv")
del data['Unnamed: 0']
data.index = [i for i in range(1, 31)]
data.head()
```

```python
sheet = from_dataframe(data)

search_box = Text(description='Search:')
link((search_box, 'value'), (sheet, 'search_token'))
VBox((search_box, sheet))
```

```python
df_alumnos_con = to_dataframe(sheet)
df_alumnos_con['per_asist'] = ((df_alumnos_con.iloc[:,1:].sum(axis=1))/32)*100
df_alumnos_con.to_csv('each_day_list_students_pooii_201.csv')
df_alumnos_con.head()
```
